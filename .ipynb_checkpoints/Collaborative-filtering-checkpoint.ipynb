```python
import pandas as pd 
import numpy as np
```

```python
movies=pd.read_csv('Movie_Id_Titles.csv')
movies = movies.rename(columns={"item_id": 'movieId'})
movies.head()
```

```python
ratings=pd.read_csv('Dataset.csv')
ratings = ratings.rename(columns={"user_id": 'userId', "item_id": 'movieId'})
ratings.head()
```

```python
final_dataset = ratings.pivot(index='movieId',columns='userId',values='rating')
final_dataset.head()
```

```python
final_dataset.fillna(0,inplace=True)
final_dataset.head()
```

```python
no_user_voted = ratings.groupby('movieId')['rating'].agg('count')
no_user_voted
```

```python
no_movies_voted = ratings.groupby('userId')['rating'].agg('count')
no_movies_voted
```

```python
final_dataset = final_dataset.loc[no_user_voted[no_user_voted > 10].index,:]
final_dataset
```

```python
final_dataset=final_dataset.loc[:,no_movies_voted[no_movies_voted > 50].index]
final_dataset
```

```python
from scipy.sparse import csr_matrix

csr_data = csr_matrix(final_dataset.values)
final_dataset.reset_index(inplace=True)
final_dataset
```

```python
from sklearn import neighbors
knn = neighbors.NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
knn.fit(csr_data)
```

```python
def get_movie_recommendation(movie_name):
    
    n_movies_to_reccomend = 10
    movie_list = movies[movies['title'].str.contains(movie_name)]  
    
    if len(movie_list):        
        movie_idx= movie_list.iloc[0]['movieId']
        movie_idx = final_dataset[final_dataset['movieId'] == movie_idx].index[0]
        distances , indices = knn.kneighbors(csr_data[movie_idx],n_neighbors=n_movies_to_reccomend+1)    
        rec_movie_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
        recommend_frame = []
    
        for val in rec_movie_indices:
            movie_idx = final_dataset.iloc[val[0]]['movieId']
            idx = movies[movies['movieId'] == movie_idx].index
            recommend_frame.append({'Title':movies.iloc[idx]['title'].values[0],'Distance':val[1]})
        
        df = pd.DataFrame(recommend_frame,index=range(1,n_movies_to_reccomend+1))
        return df
    
    else:
        return "No movies found. Please check your input"
```

```python
get_movie_recommendation('Toy Story (1995)')
```
